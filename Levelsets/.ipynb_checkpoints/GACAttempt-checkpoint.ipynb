# A Demo of the Levelset implementation of Geodesic Active Contours

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
import findiffs as fd
from redistance import redistance
import propagation_schemes as ps
from scalespace import gaussian_scalespace
```

```python
# a very simple image and a very simple level set to start with
v = np.load('scene.npy')
phi0 = np.load('phi0.npy')
plt.imshow(v, cmap='gray')
plt.contour(phi0, [0], colors=['r'], linewidths=2.5)
```

```python
# compute the edge function G
vx, vy = gaussian_scalespace(v, sigma=3.0, order=((1,0),(0,1)))
ngv2 = vx**2 + vy**2
G = 1/(1 + ngv2)
Gx = fd.cdiff_x(G)
Gy = fd.cdiff_y(G)
fig, ax = plt.subplots(1,3,sharex=True, sharey=True)
ax[0].imshow(G, cmap='gray')
ax[1].imshow(Gx, cmap='gray')
ax[2].imshow(Gy, cmap='gray')
```

```python
# Let's experiment with GAC and its different terms
def runGACIterations(G, Gx, Gy, phi, balloon, outer_its=20, inner_its=20, redist=5):
    plt.imshow(v, cmap='gray')
    # the contour structure
    cnt = plt.contour(phi, [0], colors=['r'], linewidths=2.5)
    plt.pause(0.1)
    plt.gcf().canvas.draw()
    ax = plt.gca()
    
    for j in range(outer_its):
        # make phi a signed distance function, at leat close to to 0-levelset
        phi = redistance(phi, iterations=redist)
        
        # components of the velocity. 
        # Useful to declare them if one wants to play with their effect 
        L1 = np.zeros_like(phi)
        L2 = np.zeros_like(phi)
        L3 = np.zeros_like(phi)
        
        for i in range(inner_its):
            # compute the advection speed
            L1 = ps.upwind2D_Term(phi, Gx, Gy)
            # compute the curvature dependent speed
            L2 = ps.curvature2D_Term(phi, G)
            # compute the modulated balloon force
            L3 = ps.FOSC2D_term(phi, G*balloon)
            velocicty = L1 + L2 + L3
            mveloc = np.abs(velocicty).max()
            #print('({},{}): max velocity = {}'.format(i,j,mveloc))
            # update level set
            phi = phi + (1.0/mveloc)*velocicty
            
    #plt.imshow(v, cmap='gray')
    cnt = ax.contour(phi, [0], colors=['b'], linewidths=2.5)
    return phi
    
```

```python
# and Now, use it
phi = phi0.copy()
phi = runGACIterations(G, Gx, Gy, phi, balloon=0.001, outer_its=60, inner_its=10, redist=5)
```
